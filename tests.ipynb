```python
import yfinance as yf
import pandas as pd
from dotenv import load_dotenv
import os
from binance.client import Client
import pprint
from binance.enums import *

load_dotenv()
pp = pprint.PrettyPrinter(indent=4)
```

```python
api_key = os.getenv("BINANCE_API_KEY_TEST")
api_secret = os.getenv("BINANCE_API_SECRET_TEST")
client = Client(api_key, api_secret)
```

```python
change_24h_ticker = client.futures_ticker()
change_24h_ticker #dis gives tickers that are not available for trade at futres, so
```

```python
exchange_info = client.futures_exchange_info()
print(exchange_info)
```

```python
#Check if they are still traded!
trading_symbols = {s['symbol'] for s in exchange_info['symbols'] if s['status'] == 'TRADING'}

#Filter then sort
available_tickers = [ticker for ticker in change_24h_ticker if ticker['symbol'] in trading_symbols]
sorted_tickers = sorted(available_tickers, key=lambda x: float(x['priceChangePercent']), reverse=True) #Descending

#Lets see whats up!
for idx, item in enumerate(sorted_tickers):
    if idx == 5: #Just print 5 :)
        break
    print(item['symbol'], item['priceChangePercent'], "%")
```
